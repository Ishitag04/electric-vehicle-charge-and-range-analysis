```python
import pandas as pd

df = pd.read_csv("electric_vehicles_spec_2025.csv")

df.head()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df['fast_charge_port']
```

```python
df['torque_nm'] = df['torque_nm'].fillna(df['torque_nm'].mean())
```

```python
df['model'] = df['model'].fillna(df['model'].mode)
```

```python
df['number_of_cells'] = df['number_of_cells'].fillna(df['number_of_cells'].mean())
```

```python
df['fast_charging_power_kw_dc'] = df['fast_charging_power_kw_dc'].fillna(df['fast_charging_power_kw_dc'].mean())
```

```python
df['fast_charge_port'] = df['fast_charge_port'].fillna(df['fast_charge_port'].mode)
```

```python
df['cargo_volume_l'] = df['cargo_volume_l'].fillna(df['cargo_volume_l'].mode)
```

```python
df['towing_capacity_kg'] = df['towing_capacity_kg'].fillna(df['towing_capacity_kg'].mean())
```

```python
df.isnull().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.shape
```

```python
df["efficiency_score"] = 1000 / df["efficiency_wh_per_km"]
```

```python
df["range_per_kwh"] = df["range_km"] / df["battery_capacity_kWh"]
```

```python
df["performance_index"] = df["torque_nm"] / df["acceleration_0_100_s"]
```

```python
df.to_csv("clean_ev_dataset.csv", index=False)
```
